```python
import pandas as pd

reviews = pd.DataFrame({
    "country": ["France", "France","France","Italy", "Italy"],
    "province": ["Boardeaux", "Boardeaux", "Burgundy", "Tuscany", "Tuscany"],
    "winery": ["X", "Y", "Z", "A", "B"],
    "points": [90,95,92,91,97]
})

result = reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()]).reset_index(drop=True)
result
# When grouping by two columns we get a multi-indexed DF, we can use reset_index() to bring back the normal indices
```

```python
import pandas as pd

school1 = pd.DataFrame({
    "student_id":[1,2,3],
    "names": ["Honore", "Eric", "Jesus"],
    "class": ['year1C', 'year1B', 'year1A']
})
school2 = pd.DataFrame({
    "student_id2":[4,5,6],
    "names": ["Agaba", "Don", "Justin"],
    "class": ['year1D', 'year1E', 'year1F']
})

# ? COncatnation just joins the table withou conditions
# The default concatnation direction is Row-wise, or axis=0
pd.concat([school1, school2], axis=0)
pd.concat([school1, school2], axis=1)
```

```python
import pandas as pd

school1 = pd.DataFrame({
    "student_id":[1,2,3],
    "names": ["Honore", "Eric", "Jesus"],
    "class": ['year1C', 'year1B', 'year1A']
})
school2 = pd.DataFrame({
    "student_id2":[4,5,6],
    "names": ["Agaba", "Don", "Justin"],
    "class": ['year1D', 'year1E', 'year1F']
})

# ~ LEFT JOIN(Default)

# Join joins tables based on conditions

# we need suffixes to avoid name collision when there are collumns that have the same name in both DFs
# On joinind depending in how the joint will be done , when DF's index that is going to be joined does not match with the join , the whole row is ignored and turned into NaN
# If it is a left join , means that the right Df is the one that must follow the conditions of the left Df

school1_index = school1.set_index('student_id')
school1_index.join(school2, lsuffix="_left", rsuffix="_right", how='left')

# ~ RIGHT JOIN

school2_index = school2.set_index('student_id2')
school2_index.join(school1_index, lsuffix="_left", rsuffix="_right")
```
