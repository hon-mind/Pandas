```python
# GroupBy()  ie. Split-Apply-Combine
# When you say groupBy("team"), here pandas will create a group by object where keys are the name of group and the values are the Dataframe with rows of each 
# {
#    "A": DataFrame_of_A_rows,
#    "B": DataFrame_of_B_rows
# }

import pandas as pd

team = pd.DataFrame({
    "teams":["A", "B", "A", "B", "A"],
    "points": [20,23,65,345,80],
    "marks": [34,90,87,25,45]
})
mean = team.groupby("teams").mean()
groups = team.groupby("teams")
groups.get_group("A")

teams_count = team.teams.value_counts()
teams_count2 = team.groupby('teams')["teams"].count()
teams_count2
```

```python
import pandas as pd

reviews = pd.DataFrame({
    "country": ["France", "France","France","Italy", "Italy"],
    "province": ["Boardeaux", "Boardeaux", "Burgundy", "Tuscany", "Tuscany"],
    "winery": ["X", "Y", "Z", "A", "B"],
    "points": [90,95,92,91,97]
})

result = reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()]).reset_index(drop=True)
result
# When grouping by two columns we get a multi-indexed DF, we can use reset_index() to bring back the normal indices
```

```python
import pandas as pd

reviews = pd.DataFrame({
    "country": ["France", "France","France","Italy", "Italy"],
    "province": ["Boardeaux", "Boardeaux", "Burgundy", "Tuscany", "Tuscany"],
    "winery": ["X", "Y", "Z", "A", "B"],
    "points": [90,95,92,91,97]
})

aggrigate = reviews.groupby("country")['points'].agg([min,max]).reset_index()
aggrigate
```
