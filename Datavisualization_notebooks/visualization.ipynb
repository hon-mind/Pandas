```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame(
    np.random.randint(50, 100, size=(10, 12)),
    index=[f"Row_{i}" for i in range(1, 11)],
    columns=[f"Col_{i}" for i in range(1, 13)]
)

plt.title("Heatmap Graph showing the relationship between values")
sns.heatmap(data=df, annot=True)
```
